```python
breast = sds.load_breast_cancer()
X_breast = pd.DataFrame(breast.data)
X_breast.describe()
y_breast = breast.target
print(breast.DESCR)
```

```python
X_breast.describe()
```

```python
corr_breast = X_breast.corr()

plt.figure(figsize=(20,20)) 
sns.heatmap(corr_breast, 
        xticklabels=corr_breast.columns,
        yticklabels=corr_breast.columns);
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB

bern = BernoulliNB()
multi = MultinomialNB()
gauss = GaussianNB()

print('DIGITS:')
print('bernoulli:', cross_val_score(bern, X_digs, y_digs).mean())
print('multinomial:', cross_val_score(multi, X_digs, y_digs).mean())
print('gaussian:', cross_val_score(gauss, X_digs, y_digs).mean())
print()

print('BREAST CANCER:')
print('bernoulli:', cross_val_score(bern, X_breast, y_breast).mean())
print('multinomial:', cross_val_score(multi, X_breast, y_breast).mean())
print('gaussian:', cross_val_score(gauss, X_breast, y_breast).mean())
# question 3: 3 4
```
